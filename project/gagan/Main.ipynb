## Imports

```python
import pandas as pd

from face_tagging.visualization import visualize
from face_tagging.model import FaceTaggingModel
```

## Constants

```python
TRAIN_IMAGES_DIR = "resources/train_images"
WORKFLOW_IMAGES_DIR = "resources/workflow"
TRAIN_DF_DIR = "resources/csv/key_points.csv"
```

```python
df = pd.read_csv(TRAIN_DF_DIR)
df.head()
```

```python
df = df.set_index('Unnamed: 0')
df.head(2)
```

```python
obj_model = FaceTaggingModel(keypoints_df=df, train_images_path=TRAIN_IMAGES_DIR)
```

```python
obj_model.train_model(silent=True)
```

```python
result = obj_model.matched_groups(); result
```

```python
visualize(result, TRAIN_IMAGES_DIR)
```
