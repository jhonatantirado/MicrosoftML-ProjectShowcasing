## What means explore and prepare your data ?

### Steps to follow
___

- Explore the data
   
   
- Prepare the data
   
   
#### Why explore the data ?
___

Exploring the data will enable you to better know the inputs data (independent features) and their quality:
     
 - Their **type** : numerical (discrete, continuous), categorical (numerical, ordinal, nominal)  [a little reminder if necessary ðŸ™‚](https://github.com/NAVERGONI/MicrosoftML-ProjectShowcasing/blob/master/project/C%C3%A9cile%20N/AnalyzeTheProblem.ipynb)
    
 - What are their **meaning** (e.g.: where did a passenger embark)?
    
 - Are the values all **distinct** ? Is there a limited amount of distinct values ?
    
 - Is there **missing values** ? If there is missing values is it just some or a lot ?
    
 - Is there **outliers** ?
    
 - Are some independent features **correlated** ?

#### What did we learn with this exploration and what can we do ?
___

  - There is two features for which all values are unique: PassegerId and Name => we can drop this columns
     
     
  - There is three features with missing values: 
     
     
    - Cabin: there is 687 missing values. This is too much => we will drop this column
        
    - Age: there is 177 missing values. We can decide a strategy to replace them => as this is an age we can replace them by the mean.
     
    - Embarked: there is only two missing values. As there is only three different unique values for this feature, we can replace the missing values by the most frequent value that is "S" (for Southampton)
