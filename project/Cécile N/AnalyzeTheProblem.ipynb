The goal is to predict if a passenger **did survive or not** using the known features as sex, age, ...


## What are we facing in terms of Machine Learning vocabulary ?

### The different types of approaches

There are three main approaches to machine learning :

- Supervised learning : learns from data that contains both the inputs **AND** expected outputs

- Unsupervised learning : learns from data that contains only inputs and finds hidden structures in data

- Reinforcement learning : learns how an agent should take actions in an environment to maximize a reward function

In the Titanic dataset *we know if a person survived or not* => **we have the inputs** (age, sex, ...) **AND** the **expected outputs** (survival = 0 or 1).

So we can say that our approach to this problem is a **supervised** one.

### The different types of data

### Lets dive deeper into Qualitative Data (Categorical Data)  
___

   
Below you can see the visual I created for the visual challenge in the **__Microsoft Azure ML Scholarship__**.
   
It shows that there is three types of Qualitative Data (Categorical Data) classifications:
   
- **Two-class** classification => used when the expected output is a Categorical Data that has only two distinct values  
   
   
- **Multi-class single-label** classification => used when the expected output for an instance (an item) has more than two possible output values but is described by one and only one value (eg. : the item **is** a banana)  
   
   
- **Multi-class multi-label** classification => used when the expected output for an instance (an item) has more than two possible output values and is described by one or more values (eg. : the item **contains** banana **and** peach)  
   
___
